```python
from pathlib import Path
import subprocess

source_folder = Path.cwd()
env_name = "retina_jul5"
timeout = 28800   # 8 hours
runner_name = "run_all_notebooks.ipynb"  # change this to this notebook's filename

# Set to something like "Fig 2" to only run notebooks whose names start with that text.
# Set to "" or None to run all notebooks.
name_startswith = "Supp"

continue_on_error = True

notebooks = sorted(
    p for p in source_folder.glob("*.ipynb")
    if p.name != runner_name and ".ipynb_checkpoints" not in p.parts
)

if name_startswith:
    notebooks = [p for p in notebooks if p.name.startswith(name_startswith)]

print(f"Found {len(notebooks)} notebooks")

for nb_path in notebooks:
    print(f"Running: {nb_path.name}")
    try:
        subprocess.run(
            [
                "conda", "run", "-n", env_name,
                "python", "-m", "jupyter", "nbconvert",
                "--to", "notebook",
                "--execute",
                "--inplace",
                "--ExecutePreprocessor.kernel_name=python3",
                f"--ExecutePreprocessor.timeout={timeout}",
                str(nb_path),
            ],
            check=True,
        )
        print(f"Done: {nb_path.name}\n")
    except subprocess.CalledProcessError as e:
        print(f"FAILED: {nb_path.name}\n{e}\n")
        if not continue_on_error:
            break
```
